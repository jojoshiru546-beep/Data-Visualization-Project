# Dataset Loader
Loads the dataset accordingly depending on the environment

**Google Drive folder:** https://drive.google.com/drive/folders/1MmV6mQYdxmTR9UXO5maEgoEn_R1Y2vdW

```python
import sys
import os
from pathlib import Path

IN_COLAB = 'google.colab' in sys.modules
print(f"Environment: {'Google Colab' if IN_COLAB else 'Local'}")
```

## In Google Colab
Mounts our Google Drive and reads the dataset directly from the shared folder.

```python
if IN_COLAB:
    from google.colab import drive
    drive.mount('/content/drive')

    DRIVE_PATH = Path('/content/drive/MyDrive/Data-Visualization-Project/data/raw/Global Health Statistics.csv')
    DATA_PATH = DRIVE_PATH
    print(f"Using Drive path: {DATA_PATH}")
```

## In Local Machine (VS Code on Ubuntu or Windows)
Since we use different OSs, we opted for a loader that works on whichever environment

```python
if not IN_COLAB:
    # Resolve project root regardless of whether notebook is run from /notebooks or /
    PROJECT_ROOT = Path().resolve()
    if PROJECT_ROOT.name == 'notebooks':
        PROJECT_ROOT = PROJECT_ROOT.parent

    DATA_PATH = PROJECT_ROOT / 'data' / 'raw' / 'Global Health Statistics.csv'

    # If the file is missing, download it from Google Drive automatically
    if not DATA_PATH.exists():
        print("Dataset not found locally. Downloading from Google Drive...")
        import subprocess
        subprocess.run(['pip', 'install', 'gdown', '-q'], check=True)
        import gdown
        
        # Direct file download — replace FILE_ID with the actual file ID from the Drive link
        FILE_ID = '1cug4qWE6qFArHmYwcXUdDMaJIfmUzAZD'
        gdown.download(f'https://drive.google.com/uc?id={FILE_ID}', str(DATA_PATH), quiet=False)
    else:
        print(f"Dataset found: {DATA_PATH}")
```

## Load & Verify

```python
import pandas as pd

df = pd.read_csv(DATA_PATH)

print(f"Shape      : {df.shape[0]:,} rows × {df.shape[1]} columns")
print(f"Columns    : {df.columns.tolist()}")
print(f"Memory     : {df.memory_usage(deep=True).sum() / 1e6:.1f} MB")
df.head(3)
```

```python
# Quick No-Errors Check
assert df.shape[0] > 0, "Dataset is empty!"
assert 'Country' in df.columns, "Expected column 'Country' not found!"
assert 'Disease Name' in df.columns, "Expected column 'Disease Name' not found!"
print("All checks passed. Dataset is ready.")
```
